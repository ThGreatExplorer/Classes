```python
import numpy as np
import matplotlib.pyplot as plt
```

# Using Matplotlib

```python
fig, ax = plt.subplots()
x = np.linspace(-5,5,100) # creates evenly space elemeents in closed interval from [-5,5]

def y(x):
    return x ** 2

ax.plot(x, y(x))
plt.show()
```

```python
# working with 3d plots
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

x = np.linspace(-100,100,5)
y = np.linspace(-100,100,5)

x_mesh, y_mesh = np.meshgrid(x,y)
print(x_mesh)
print(y_mesh)
print("""This is essentially constructing each possible (x,y) pair: 
(-100, -100), (-50, -100),  (0, -100),  (50, -100),  (100, -100)
(-100, -50),  (-50, -50),   (0, -50),   (50, -50),   (100, -50)
(-100, 0),    (-50, 0),     (0, 0),     (50, 0),     (100, 0)
(-100, 50),   (-50, 50),    (0, 50),    (50, 50),    (100, 50)
(-100, 100),  (-50, 100),   (0, 100),   (50, 100),   (100, 100)
""")

def z(x,y):
    return (x-y)**2

ax.plot_surface(x_mesh, y_mesh, z(x_mesh,y_mesh), cmap="viridis")
plt.show()
```

```python
# working with scatterplots

fig = plt.figure()
x = np.linspace(0,10,100)
y = 5 * x + 2 + np.random.normal(0,4,100) # mean = 0, variance = 4, 100 samples drawn so 1x100 vector 

plt.plot(x,y, '.')
plt.show()
```

```python
fig = plt.figure()
def mse(m,c):
    return np.mean((y - (m * x + c))**2)

# plot the loss function
m = np.linspace(-100,100,100) # samples possible values for the mean between -100, 100
c = np.linspace(-100,100,100) # samples possible values for the intercept between -100, 100
m_mesh, c_mesh = np.meshgrid(m,c)
ax = fig.add_subplot(projection='3d')
mse_loss = np.zeros((100,100)) # each i,j represents the mse loss for the graph above with slope = m[i,j] and intercept = c[i,j] 

for i in range(100):
    for j in range(100):
        mse_loss[i,j] = mse(m_mesh[i,j], c_mesh[i,j])

print(f"Minimum Cost: np.min(mse_loss)")
m_idx, c_idx = np.unravel_index(np.argmin(mse_loss), mse_loss.shape)
print(f"Optimal slope (m): {m_mesh[m_idx,c_idx]}")
print(f"Optimal intercept (c): {c_mesh[m_idx,c_idx]}")

        
ax.plot_surface(m_mesh, c_mesh, mse_loss, cmap='viridis')
ax.plot(m_mesh[m_idx,c_idx],c_mesh[m_idx,c_idx], 'x', color="black")
plt.show()
```
